```python
from unsloth import FastModel

model, tokenizer = FastModel.from_pretrained(
    model_name="Qwen/Qwen2.5-3B-Instruct",
    max_seq_length=2048,
    load_in_4bit=True,
)

print(" Model Loaded Successfully!")
```

```python
from unsloth import FastModel

model = FastModel.get_peft_model(
    model,
    r=16,
    target_modules=[
        "q_proj",
        "k_proj",
        "v_proj",
        "o_proj",
        "gate_proj",
        "up_proj",
        "down_proj",
    ],
    lora_alpha=16,
    lora_dropout=0,
    bias="none",
    use_gradient_checkpointing="unsloth",
    random_state=3407,
)
```

```python
import json

records = []

decoder = json.JSONDecoder()

with open("/content/Chat Data for assessment of applicants.json", "r", encoding="utf-8") as f:
    text = f.read()

idx = 0
length = len(text)

while idx < length:
    while idx < length and text[idx] in " \n\r\t,":
        idx += 1

    if idx >= length:
        break

    try:
        obj, end = decoder.raw_decode(text, idx)
        records.append(obj)
        idx = end
    except json.JSONDecodeError:
        idx += 1

print("Total conversations:", len(records))
print(records[0].keys())
```

```python
from datasets import Dataset

dataset = Dataset.from_list(records)

print(dataset)
print(dataset[0])
```

```python
def formatting_prompts_func(examples):
    texts = []
    for messages in examples["messages"]:
        text = tokenizer.apply_chat_template(
            messages,
            tokenize=False,
            add_generation_prompt=False,
        )
        texts.append(text)
    return {"text": texts}

dataset = dataset.map(
    formatting_prompts_func,
    batched=True,
)

print(dataset[0]["text"][:500])
```

```python
from trl import SFTTrainer, SFTConfig

trainer = SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    train_dataset=dataset,
    args=SFTConfig(
        output_dir="outputs",
        max_seq_length=2048,
        per_device_train_batch_size=2,
        gradient_accumulation_steps=4,
        learning_rate=2e-4,
        num_train_epochs=3,
        logging_steps=1,
        save_strategy="epoch",
        fp16=True,
    ),
    dataset_text_field="text",
)
```

```python
trainer.train()
```

```python
model.save_pretrained("vedaz_qwen_lora")
tokenizer.save_pretrained("vedaz_qwen_lora")

print(" LoRA adapter saved!")
```

```python
from unsloth import FastModel

FastModel.for_inference(model)

messages = [
    {
        "role": "user",
        "content": "My DOB is 12 May 1998, 6:30 AM, Hyderabad. Will I get a job this year?"
    }
]

inputs = tokenizer.apply_chat_template(
    messages,
    tokenize=True,
    return_tensors="pt",
    add_generation_prompt=True,
).to(model.device)

outputs = model.generate(
    input_ids=inputs,
    max_new_tokens=256,
    temperature=0.7,
)

print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```

```python
from unsloth import FastModel

FastModel.for_inference(model)

messages = [
    {
        "role": "system",
        "content": "You are Vedaz's AI Vedic astrologer. You give compassionate, balanced, non-fatalistic guidance."
    },
    {
        "role": "user",
        "content": "My DOB is 12 May 1998, 6:30 AM, Hyderabad. Will I get a job this year?"
    }
]

inputs = tokenizer.apply_chat_template(
    messages,
    tokenize=True,
    add_generation_prompt=True,
    return_tensors="pt",
)

inputs = {k: v.to(model.device) for k, v in inputs.items()} if isinstance(inputs, dict) else inputs.to(model.device)

outputs = model.generate(
    **inputs if isinstance(inputs, dict) else {"input_ids": inputs},
    max_new_tokens=256,
    temperature=0.7,
    do_sample=True,
)

print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```
